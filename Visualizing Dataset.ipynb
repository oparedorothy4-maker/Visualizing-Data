```python
# Import all necessary libraries:
# pandas for data manipulation and analysis
# matplotlib.pyplot for creating static plots and visualizations
# seaborn for creating advanced statistical plots with better aesthetics
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the CSV file "teen_phone_addiction_dataset.csv" into a Pandas DataFrame called df
df = pd.read_csv("teen_phone_addiction_dataset.csv")

# Preview the first five rows of the DataFrame to get an initial look at the data
df.head()
```

```python
# Display the dataset structure, including column names, data types, and non-null counts to check for missing values
df.info()
```

```python
# Describing the Location (Mean, Median, Mode for each variable)

# Daily Usage Hours
mean_usage = df["Daily_Usage_Hours"].mean()       # Calculate the average daily usage hours
median_usage = df["Daily_Usage_Hours"].median()   # Calculate the median daily usage hours
mode_usage = df["Daily_Usage_Hours"].mode()[0]    # Calculate the mode (most frequent value) of daily usage hours
print(mean_usage)
print(median_usage)
print(mode_usage)

# Sleep Hours
mean_sleep = df["Sleep_Hours"].mean()             # Calculate the average sleep hours
median_sleep = df["Sleep_Hours"].median()         # Calculate the median sleep hours
mode_sleep = df["Sleep_Hours"].mode()[0]          # Calculate the mode of sleep hours
print(mean_sleep)
print(median_sleep)
print(mode_sleep)

# Addiction Level
mean_addiction = df["Addiction_Level"].mean()     # Calculate the average addiction level
median_addiction = df["Addiction_Level"].median() # Calculate the median addiction level
mode_addiction = df["Addiction_Level"].mode()[0]  # Calculate the mode of addiction level
print(mean_addiction)
print(median_addiction)
print(mode_addiction)
```

```python
# Describing the Variance of each variable

# Calculate the variance of Daily Usage Hours
variance_usage = df["Daily_Usage_Hours"].var()

# Calculate the variance of Sleep Hours
variance_sleep = df["Sleep_Hours"].var()

# Calculate the variance of Addiction Level
variance_addiction = df["Addiction_Level"].var()

# Print the variances
print(variance_usage)
print(variance_sleep)
print(variance_addiction)
```

```python
# Get summary statistics (count, mean, std, min, 25%, 50%, 75%, max) for the key columns
print(df[["Daily_Usage_Hours", "Sleep_Hours", "Addiction_Level"]].describe())
```

```python
# Calculate and display the covariance matrix for the key columns
# Covariance shows how two variables vary together
cov_matrix = df[["Daily_Usage_Hours", "Sleep_Hours", "Addiction_Level"]].cov()
print(cov_matrix)
```

```python
# Calculate and display the correlation matrix for the key columns
# Correlation measures the strength and direction of the linear relationship between variables
corr_matrix = df[["Daily_Usage_Hours", "Sleep_Hours", "Addiction_Level"]].corr()
print(corr_matrix)
```

```python
# Focus on the key three columns and store them in a new DataFrame called data
data = df[["Daily_Usage_Hours", "Sleep_Hours", "Addiction_Level"]]

# 1. HISTOGRAMS
# Create histograms with KDE for each key column to visualize their distributions

plt.figure(figsize=(12, 4))  # Set the figure size

for i, col in enumerate(data.columns, 1):
    plt.subplot(1, 3, i)  # Create a subplot for each column
    sns.histplot(data[col], bins=20, kde=True, color="skyblue")  # Plot histogram with KDE
    plt.title(f"Distribution of {col}")  # Set the title for each subplot

# Save all histograms as a single high-resolution image
plt.savefig("all_histograms.png", dpi=300)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plots
```

```python
# 2. SCATTER PLOTS
# Create scatter plots with regression lines to explore relationships between variables

plt.figure(figsize=(10, 4))  # Set the figure size

# Daily Usage vs Sleep Hours
plt.subplot(1, 2, 1)  # First subplot
sns.regplot(x="Daily_Usage_Hours", y="Sleep_Hours", data=data, scatter_kws={"alpha": 0.5})
plt.title("Daily Usage vs Sleep Hours")  # Set subplot title

# Daily Usage vs Addiction Level
plt.subplot(1, 2, 2)  # Second subplot
sns.regplot(x="Daily_Usage_Hours", y="Addiction_Level", data=data, scatter_kws={"alpha": 0.5}, color="orange")
plt.title("Daily Usage vs Addiction Level")  # Set subplot title

# Save all scatter plots as a single high-resolution image
plt.savefig("all_scatterplots.png", dpi=300)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plots
```

```python
# 3. CORRELATION HEATMAP
# Create a heatmap to visualize correlations between key variables

plt.figure(figsize=(8, 6))  # Set the figure size

corr = data.corr()  # Calculate the correlation matrix

# Plot the heatmap with annotations and a color map
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")  # Set the title

# Save the heatmap as a high-resolution image
plt.savefig("correlation_heatmap.png", dpi=300)

plt.show()  # Display the heatmap

```

```python
# 4. BOX PLOT
# Create a box plot to visualize the distribution and detect outliers for key variables

plt.figure(figsize=(8, 6))  # Set the figure size

data.boxplot()  # Generate the box plot for all key columns
plt.title("Box Plot of Daily Usage, Sleep Hours, and Addiction Level")  # Set the plot title
plt.ylabel("Values")  # Label the y-axis
plt.xticks(rotation=0)  # Keep x-axis labels horizontal

# Save the box plot as a high-resolution image
plt.savefig("boxplot.png", dpi=300)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the box plot
```
