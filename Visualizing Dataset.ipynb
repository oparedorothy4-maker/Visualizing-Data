```python
#Calculating the Location
# Daily Usage Hours
mean_usage = df['Daily_Usage_Hours'].mean()
median_usage = df['Daily_Usage_Hours'].median()
mode_usage = df['Daily_Usage_Hours'].mode()[0]
print(mean_usage)
print(median_usage)
print(mode_usage)

# Sleep Hours
mean_sleep = df['Sleep_Hours'].mean()
median_sleep = df['Sleep_Hours'].median()
mode_sleep = df['Sleep_Hours'].mode()[0]
print(mean_sleep)
print(median_sleep)
print(mode_sleep)

# Addiction Level
mean_addiction = df['Addiction_Level'].mean()
median_addiction = df['Addiction_Level'].median()
mode_addiction = df['Addiction_Level'].mode()[0]
print(mean_addiction)
print(median_addiction)
print(mode_addiction)
```

```python
# Calculate Variance
variance_usage = df["Daily_Usage_Hours"].var()
variance_sleep = df["Sleep_Hours"].var()
variance_addiction =df["Addiction_Level"].var()

print(variance_usage)
print(variance_sleep)
print(variance_addiction)
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("teen_phone_addiction_dataset.csv")

# Focus on the key three columns
data = df[['Daily_Usage_Hours', 'Sleep_Hours', 'Addiction_Level']]

# 1. HISTOGRAMS

plt.figure(figsize=(12, 4))
for i, col in enumerate(data.columns, 1):
    plt.subplot(1, 3, i)
    sns.histplot(data[col], bins=20, kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')

 # Save as one image
plt.savefig('all_histograms.png', dpi=300) 
    
plt.tight_layout()
plt.show()


# 2. SCATTER PLOTS

plt.figure(figsize=(10, 4))

# Daily Usage vs Sleep Hours
plt.subplot(1, 2, 1)
sns.regplot(x='Daily_Usage_Hours', y='Sleep_Hours', data=data, scatter_kws={'alpha':0.5})
plt.title('Daily Usage vs Sleep Hours')

# Daily Usage vs Addiction Level
plt.subplot(1, 2, 2)
sns.regplot(x='Daily_Usage_Hours', y='Addiction_Level', data=data, scatter_kws={'alpha':0.5}, color='orange')
plt.title('Daily Usage vs Addiction Level')

plt.savefig('all_scatterplots.png', dpi=300)  # save as one image

plt.tight_layout()
plt.show()

# 3. CORRELATION HEATMAP

plt.figure(figsize=(8, 6))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.savefig('correlation_heatmap.png', dpi=300)

plt.show()

# 4. BOX PLOT

plt.figure(figsize=(8, 6))
data.boxplot()
plt.title("Box Plot of Daily Usage, Sleep Hours, and Addiction Level")
plt.ylabel("Values")
plt.xticks(rotation=0)
plt.show()


```
